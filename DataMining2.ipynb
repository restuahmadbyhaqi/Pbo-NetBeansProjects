```
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
file_path='Dataset HR.xlsx'
df = pd.read_excel(file_path)
print(df.head(10))
```

```
print(df.dtypes)
```

```
print(df.info)
```

```
print(df.describe)
```

```
plt.figure(figsize=(10,5))


df['sales'].value_counts().plot.bar(color='skyblue')

plt.title('Distribusi Penjualan per Kategori')
plt.xlabel('Kategori Sales')
plt.ylabel('Jumlah')

plt.show()
```

```
plt.boxplot(df['time_spend_company'])
plt.show()
```

```
plt.boxplot(df['satisfaction_level'])
plt.show()
```

```

df_numeric = df.select_dtypes(include=['float64', 'int64'])
plt.figure(figsize=(10, 10))
correlation_matrix = df_numeric.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap Korelasi Antar Variabel')
plt.show()
```
